## 1. Why Variables Matter in Data Engineering

In a real data pipeline, Python stores values such as:
- order amount
- user name
- transaction status
- row count
- validation result

A variable is just a name we give to a value so that we can use it again in our program.


```python
# Examples of variables in a data engineering context
order_id = 101
order_amount = 299.99
customer_name = "Anuj"
is_valid_order = True

print(order_id)
print(order_amount)
print(customer_name)
print(is_valid_order)
```

## 2. Dynamic Typing

Python is **dynamically typed**.

That means:
- You do **not** need to declare the type manually.
- Python decides the type at runtime based on the value assigned.

For example, the same variable name can point to different types at different times.


```python
x = 10
print(x, type(x))

x = 10.5
print(x, type(x))

x = "ten"
print(x, type(x))

x = False
print(x, type(x))
```

## 3. Memory Management (Intro)

In Python, a variable name refers to an object in memory.

Two useful functions:
- `type()` tells us the data type.
- `id()` gives us the identity of the object in memory.

We will keep this at a basic level for now.


```python
a = 105
b = 105
c = 101
d = 100
e = "Ram"
f = "Ram"
g = "RAM"
print("a:", a, "type:", type(a), "id:", id(a))
print("b:", b, "type:", type(b), "id:", id(b))
print("c:", c, "type:", type(c), "id:", id(c))
print("d:", d, "type:", type(d), "id:", id(d))
print("e:", e, "type:", type(e), "id:", id(e))
print("f:", f, "type:", type(f), "id:", id(f))
print("g:", g, "type:", type(g), "id:", id(g))
```

### Note
If two variables refer to the same object, their `id()` may be the same.

```python
name = "Data Engineering"
print("name:", name)
print("type:", type(name))
print("id:", id(name))
```

## 4. Numeric Types

Python has three important numeric types for this lesson:

### `int`
Whole numbers.

### `float`
Decimal numbers.

### `complex`
Numbers with a real and imaginary part.

Complex numbers are not commonly used in everyday data engineering, but students should know they exist.


```python
row_count = 2500          # int
average_price = 149.75   # float
signal1 = 3 + 4j         # complex
signal2 = 3 + 4j

print(row_count, type(row_count))
print(average_price, type(average_price))
print(signal1, type(signal1))
print(signal1 + signal2, type(signal1))
```

```python
# Numeric operations
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulo:", a % b)
print("Power:", a ** b)
```

## 5. Boolean Logic

The `bool` type has only two values:
- `True`
- `False`

Booleans are extremely important in data engineering because filtering and validation logic rely on them.


```python
is_amount_positive = True
is_status_completed = False

print(is_amount_positive, type(is_amount_positive))
print(is_status_completed, type(is_status_completed))
```

```python
# Comparison expressions return booleans
amount = 250
print(amount > 0)
print(amount == 250)
print(amount < 0)
```

```python
# Truth values of different data
print(bool(0))
print(bool(1))
print(bool(""))
print(bool("data"))
print(bool([]))
print(bool([1, 2, 4,6]))
```

## 6. Type Casting

Type casting means converting data from one type to another.

Common conversions:
- `int()`
- `float()`
- `str()`


```python
price_text = "199.99"
price_number = float(price_text)

print(price_text, type(price_text))
print(price_number, type(price_number))
```

```python
user_age = "25"
user_age_int = int(user_age)

print(user_age, type(user_age))
print(user_age_int, type(user_age_int))
```

```python
order_count = 50
order_count_text = str(order_count)

print(order_count, type(order_count))
print(order_count_text, type(order_count_text))
```

## 7. Dynamic Assignment — Swapping Without a Third Variable

Python allows clean swapping in one line.


```python
x = 10
y = 2000

print("Before swap:", x, y)
x, y = y, x
print("After swap:", x, y)
```

## 8. Exploring with `type()` and `id()`

This is one of the most important practical pieces of today’s class.
Use these functions to inspect variables and observe how Python behaves.


```python
sample_value = 42.25
print("Value:", sample_value)
print("Type:", type(sample_value))
print("ID:", id(sample_value))

sample_value = "forty two"
print("\nAfter reassignment:")
print("Value:", sample_value)
print("Type:", type(sample_value))
print("ID:", id(sample_value))
```

## 9. Mini Data Engineering Example

Let’s connect today’s topic to the running course project.

We receive some raw order-like values. We inspect their types, convert them where needed, and validate them.


```python
raw_order_id = "1001"
raw_amount = "450.75"
raw_customer = "Riya"
raw_status = "True"

print(raw_order_id, type(raw_order_id))
print(raw_amount, type(raw_amount))
print(raw_customer, type(raw_customer))
print(raw_status, type(raw_status))
```

```python
# Convert raw values into useful Python types
order_id = int(raw_order_id)
amount = float(raw_amount)
customer = str(raw_customer)

# Since raw_status is a string, let's manually interpret it
is_valid = raw_status == "True"

print(order_id, type(order_id))
print(amount, type(amount))
print(customer, type(customer))
print(is_valid, type(is_valid))
```

```python
# Validation logic using boolean expressions
is_amount_positive = amount > 0
is_order_usable = is_valid and is_amount_positive

print("Amount positive:", is_amount_positive)
print("Order usable:", is_order_usable)
```
