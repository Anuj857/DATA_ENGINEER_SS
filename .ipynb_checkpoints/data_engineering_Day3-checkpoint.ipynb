## use case: trading platform example

```python
import random
import time
def generate_trade():
    symbols = ["AAPL", "TSLA", "GOOG"]
    sides = ["BUY", "SELL"]

    trade = {
        "trade_id": random.randint(1000, 9999),
        "symbol": random.choice(symbols),
        "side": random.choice(sides),
        "quantity": str(random.randint(1, 100)),
        "price": str(round(random.uniform(100, 500), 2))
    }

    return trade
```

```python
#print unlime
import time
while True:
  trade = generate_trade()
  print(trade)
  time.sleep(2)
```

USE the generate_trade method to receive trades and process the trades for next 1 minute to find out the P&L

```python
import time

start_time = time.time()
duration = 60

total_pnl = 0
processed_trades = 0

print("Starting trade processing for 1 minute...")

while (time.time() - start_time) < duration:
    trade = generate_trade()

    # Convert quantity and price to appropriate types
    quantity = int(trade['quantity'])
    price = float(trade['price'])

    if trade['side'] == 'BUY':
        total_pnl -= (quantity * price)
    elif trade['side'] == 'SELL':
        total_pnl += (quantity * price)

    processed_trades += 1
    time.sleep(1) # Small delay to avoid busy-waiting and allow for other operations

print(f"\nFinished processing trades. Total trades processed: {processed_trades}")
print(f"Total P&L after 1 minute: ${total_pnl:,.2f}")
```

Q. print first 10 user-email who have age >25 and gender ="female"

```python
target_customers = []
total_users = 0
while len(target_customers) < 10:
  response = requests.get("https://randomuser.me/api/")
  users = response.json()["results"]
  if(len(users) > 0):
    total_users+=1
    user = users[0]
    if(user['gender'] == 'female' and user['dob']['age'] > 35):
      target_customers.append(user["email"])
print(target_customers, total_users)
```
